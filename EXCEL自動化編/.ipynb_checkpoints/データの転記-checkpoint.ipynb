```python
import openpyxl

wb=openpyxl.load_workbook("請求リスト.xlsx")
ws=wb["Sheet1"]

wb_test=openpyxl.load_workbook("請求書_テスト１.xlsx", data_only=True)
ws_test=wb_test.worksheets[0]

ws.cell(2, 1).value=ws_test.cell(3, 14).value
ws.cell(2, 2).value=ws_test.cell(3, 1).value
ws.cell(2, 3).number_format=r"¥#,##0;¥-#,##0"
ws.cell(2, 3).value=ws_test.cell(15, 4).value
ws.cell(2, 4).number_format=r"yyyy/m/d"
ws.cell(2, 4).value=ws_test.cell(4, 14).value

wb.save("請求リスト.xlsx")
```

```python
# 完成版

import openpyxl
from glob import glob

wb=openpyxl.load_workbook("請求リスト.xlsx")
ws=wb["Sheet1"]

files=sorted(glob("請求書*.xlsx"))

for file in files:
    wb_test=openpyxl.load_workbook(file, data_only=True)
    ws_test=wb_test.worksheets[0]

    input_row=ws.max_row+1
    ws.cell(input_row, 1).value=ws_test.cell(3, 14).value
    ws.cell(input_row, 2).value=ws_test.cell(3, 1).value
    ws.cell(input_row, 3).number_format=r"¥#,##0;¥-#,##0"
    ws.cell(input_row, 3).value=ws_test.cell(15, 4).value
    ws.cell(input_row, 4).number_format=r"yyyy/m/d"
    ws.cell(input_row, 4).value=ws_test.cell(4, 14).value

wb.save("請求リスト.xlsx")
```

```python
# 請求リストのテンプレートを自作する

import openpyxl
from openpyxl.styles import Font, PatternFill, Alignment
from openpyxl.styles.borders import Border, Side

wb2=openpyxl.Workbook()
ws2=wb2.active
ws2.title="Sheet1"

header=["請求No.", "請求先会社名", "合計金額", "請求日"]
column=["A", "B", "C", "D"]

side=Side(style="thin", color="000000")
border=Border(top=side, bottom=side, left=side, right=side)

for i in range(len(header)):
    ws2.cell(1,i+1).value=header[i]
    ws2.cell(1,i+1).font=Font(bold=True)
    ws2.cell(1,i+1).fill=PatternFill(patternType="solid", fgColor="FFA500")
    ws2.cell(1,i+1).alignment=Alignment(horizontal="centerContinuous")
    ws2.cell(1,i+1).border=border
    ws2.column_dimensions[column[i]].width = 20

wb2.save("請求リスト_自作テンプレート.xlsx")
```
