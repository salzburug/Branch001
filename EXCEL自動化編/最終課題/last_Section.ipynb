```python
# 完成版
import pandas as pd

month_list=["4月","5月","6月"]

df_concat=pd.DataFrame()

for month in month_list:
    df_m=pd.read_excel("月別売上.xlsx",sheet_name=month)
    df_concat=pd.concat([df_concat,df_m])

df_concat["売上金額"]=df_concat["単価"]*df_concat["個数"]

df_s=pd.read_excel("社員リスト.xlsx",sheet_name="リスト")

df_s["社員番号"]=df_s["社員番号"].str.replace("A-","").str.replace("B-","")

df_s=df_s.astype({"社員番号":int})

df_merge=pd.merge(df_concat,df_s,on="社員番号")

name_list=df_merge["名前"].unique()

for name in name_list:
    df_each=df_merge[df_merge["名前"]==name]
    df_output=df_each.pivot_table(index="取引先",columns="商品名",values="売上金額",aggfunc="sum",fill_value=0,margins=True,margins_name="総計")
    df_output.to_excel("総売上内訳_{}.xlsx".format(name))
```
