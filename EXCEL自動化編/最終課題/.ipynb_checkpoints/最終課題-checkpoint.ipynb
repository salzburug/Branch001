```python
!pip install pandas
import pandas as pd

month_list=["4月","5月","6月"]

df_concat=pd.DataFrame()
for month in month_list:
    df_month=pd.read_excel("月別売上.xlsx",sheet_name=month)
    df_concat=pd.concat([df_concat,df_month])

df_concat
```

```python
# ②売上金額の列を追加

df_concat["売上金額"]=df_concat["単価"]*df_concat["個数"]
df_concat
```
