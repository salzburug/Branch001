# ファイルの結合

```python
from glob import glob

files=glob("売上一覧_*.xlsx")
files
```

```python
import pandas as pd

df_tanaka=pd.read_excel(files[0], sheet_name="Sheet1")
df_satou=pd.read_excel(files[1], sheet_name="Sheet1")

df_concat=pd.concat([df_tanaka, df_satou])
df_concat
```

```python
# 完成版

from glob import glob
import pandas as pd

files=glob("売上一覧_*.xlsx")

df_concat=pd.DataFrame()

for file in files:
    df_tantou=pd.read_excel(file, sheet_name="Sheet1")
    df_concat=pd.concat([df_concat, df_tantou])
    
df_concat.sort_values("No.").to_excel("売上一覧_結合.xlsx", index=False)
```

```python
with pd.ExcelWriter("売上一覧(改).xlsx", mode="a") as writer:
    df_tanaka.to_excel(writer, sheet_name="田中_{}円".format(df_tanaka["売上金額"].sum()), index=False)
```

```python
# 完成版

import pandas as pd

df=pd.read_excel("売上一覧(改).xlsx", sheet_name="まとめ", header=2)
df=df.iloc[0:len(df)-2, 1:9].dropna()

tantou_list=df["担当者"].unique()

for tantou in tantou_list:
    df_tantou=df[df["担当者"]==tantou]

    with pd.ExcelWriter("売上一覧(改).xlsx", mode="a") as writer:
        df_tantou.to_excel(writer, sheet_name="{}_{}円".format(tantou, df_tantou["売上金額"].sum()), index=False)
```
